```python
# Pie chart for smartphone market share

import matplotlib.pyplot as plt

# Data
brands = ["Samsung", "Apple", "Xiaomi", "Oppo", "Vivo"]
market_share = [30, 25, 20, 15, 10]

# Find index of maximum share
max_index = market_share.index(max(market_share))

# Explode only the highest value
explode = [0, 0, 0, 0, 0]
explode[max_index] = 0.1   # Highlight max share

# Plot pie chart
plt.pie(market_share, labels=brands, explode=explode, autopct='%1.1f%%')

# Add title
plt.title("Smartphone Market Share")  
# Output: Pie chart with Samsung highlighted (largest share)

# Show chart
plt.show()  
# Output: Pie chart displayed
```
