```python
# Line chart for monthly sales data

import matplotlib.pyplot as plt

# Monthly sales data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [12000, 15000, 13000, 18000, 20000, 17000]

# Plot line chart
plt.plot(months, sales, marker='o', label="Sales")

# Add title and labels
plt.title("Monthly Sales Data")      # Output: Title displayed on top
plt.xlabel("Months")                # Output: X-axis labeled as Months
plt.ylabel("Sales (₹)")             # Output: Y-axis labeled as Sales

# Add legend
plt.legend()                        # Output: Shows legend "Sales"

# Add grid
plt.grid()                          # Output: Grid lines appear

# Show chart
plt.show()                          # Output: Line chart displayed
```
